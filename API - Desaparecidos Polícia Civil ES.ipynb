```python
import requests
import pandas as pd
```

```python
def fetch_data(api_key, max_page, page_size):
    base_url = "https://api.sesp.es.gov.br/api-desaparecidos/v1/consultaDesaparecidos"
    data = []

    print("Iniciando a coleta de dados...")

    # Iterar pelas páginas de 0 a 19
    for page in range(max_page + 1):
        print(f"Buscando dados da página {page}...")
        params = {
            'api_key': api_key,
            'page': page,
            'size': page_size
        }
        response = requests.get(base_url, params=params)
        
        # Verifica se a resposta é bem-sucedida
        if response.status_code == 200:
            # Adiciona os dados da página atual à lista de dados
            page_data = response.json()
            data.extend(page_data['content'])  # Atualizado para usar a chave correta
            print(f"Dados da página {page} coletados com sucesso.")
        else:
            print(f"Erro ao buscar dados da página {page}: {response.status_code}")

    print("Coleta de dados completa.")
    return data

# Use esta função para carregar os dados
api_key = 'RGJcEgFZcfKiBsb5bpvT9KZQ7UeLoUkY1Ia6dXCw'
data = fetch_data(api_key, 20, 50)
```

```python
for item in data:
    item.pop('imagem', None)
```

```python
nomes = []
locais_desaparecimento = []
idades = []

for item in data:
    # Extrai o nome e adiciona à lista de nomes
    nomes.append(item.get('nome', 'Nome não informado'))  # Usa um valor padrão caso 'nome' não esteja presente

    # Extrai o local de desaparecimento e adiciona à lista de locais de desaparecimento
    locais_desaparecimento.append(item.get('cidadeFato', 'Local não informado'))  # Usa um valor padrão caso 'cidadeFato' não esteja presente

    # Extrai a idade e adiciona à lista de idades
    idades.append(item.get('idade', 'Idade não informada'))  # Usa um valor padrão caso 'idade' não esteja presente

# Exibe as listas criadas
print("Nomes:", nomes)
print("Locais de Desaparecimento:", locais_desaparecimento)
print("Idades:", idades)
```

```python
data = {
    'Nomes': nomes,
    'Idades': idades,
    'Local de Desaparecimento': locais_desaparecimento
}

df = pd.DataFrame(data)
df.to_csv('Desaparecidos Polícia Civil ES.csv', index=False)
```
