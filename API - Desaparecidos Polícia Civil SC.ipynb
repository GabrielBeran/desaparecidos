```python
import requests
import pandas as pd
```

```python
# URL da API
url = "https://devs.pc.sc.gov.br/servicos/desaparecidos/dados?pagina=0&size=1000"

# Realizando a requisição GET
response = requests.get(url)

# Verifica se a requisição foi bem-sucedida
if response.status_code == 200:
    # Convertendo a resposta em JSON
    dados = response.json()
    # Agora 'dados' é uma variável que contém todos os dados retornados pela API
else:
    print("Falha na requisição:", response.status_code)
```

```python
nomes = []
local_desaparecimento = []

# Loop para extrair os dados
for entrada in dados:
    # Extrai o nome e adiciona na lista 'nomes'
    nomes.append(entrada['nomePessoa'])
    
    # Extrai o município e adiciona na lista 'local_desaparecimento'
    local_desaparecimento.append(entrada['municipio'])

# Agora 'nomes' e 'local_desaparecimento' contêm os dados extraídos
print("Nomes:", nomes)
print("Locais de Desaparecimento:", local_desaparecimento)
```

```python
data = {
    'Nomes': nomes,
    'Local de Desaparecimento': local_desaparecimento
}

df = pd.DataFrame(data)
df.to_csv('Desaparecidos Polícia Civil SC.csv', index=False)
```
